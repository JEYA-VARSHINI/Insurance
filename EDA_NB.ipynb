# EXPLORATARY DATA ANALYSING

#### Copy

```python
dataf1=data.copy()
```

# PLOTS

#### Histogram

```python
plt.hist(dataf1['age'],edgecolor='white',color='#0e6655')
plt.title('INSURANCE')
plt.xlabel('Age')
plt.ylabel('customers')
```

##### The majority of insurance customers are in their early 20s, with the number decreasing in older age groups

#### Pie chart

```python
labels = ['male','female']
sizes = [33.8, 31.9]  # Percentages or values
colors = ['#ff9999','#66b3ff']  # Custom colors
explode = (0.1, 0)  # Explode first slice (i.e. 'male')

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
```

##### The insurance customers are nearly evenly split by gender, with 51.4% male and 48.6% female.

#### Correlation

```python
correlation_matrix = data[['age','children', 'expenses']].corr()
print('Correlation matrix:')
print(correlation_matrix)
```

#### Heat map

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', linewidths=0.5)
colors=('#85c1e9','#1b4f72','#d6eaf8')
```

##### There is a moderate positive correlation between age and expenses

#### Line plot

```python
sns.lineplot(data=data, x='bmi',y='region',color='#94b8b8')
plt.show
```

#### It shows a high concentration of BMI values between 20 and 40 across all regions, with no significant variation in BMI distribution between the regions.

#### count plot

```python
sns.countplot(x='children',data=data, color= '#f5a478')
```

##### The majority of the observations in the dataset have no children, with a significant drop-off as the number of children increases.

#### Box plot

```python
sns.boxplot(x=dataf1['bmi'],color='#ab7fdf')
```

##### The majority of BMI values are concentrated between 25 and 35, with a few outliers extending beyond 40.

#### Bar plot

```python
sns.barplot(x='age', y='sex', data=dataf1,palette='muted',orient='h')
```

##### The average age of females is slightly higher than that of males.
